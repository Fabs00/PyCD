```python
import re
def tokens(txt, idioma = 'spanish'):
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())

    stop_words = set(stopwords.words(idioma))
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado
```

```python
# Importar las librerías necesarias
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import download

# Descargar los recursos necesarios de NLTK
download('stopwords')
download('punkt')

# Función para procesar dos textos a la vez
def tokens_dos_textos(t, m, idioma='english'):
    # Verificar que las stopwords para el idioma estén disponibles
    try:
        stop_words = set(stopwords.words(idioma))
    except LookupError:
        raise ValueError(f"Stopwords para el idioma '{idioma}' no están disponibles. Asegúrate de que estén instaladas.")

    # Función interna para limpiar y filtrar texto
    def procesar_texto(txt):
        texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())
        tokens = word_tokenize(texto_limpio)
        return [word for word in tokens if word not in stop_words]

    # Procesar ambos textos
    texto_filtrado_m = procesar_texto(t)
    texto_filtrado_t = procesar_texto(m)

    return texto_filtrado_m, texto_filtrado_t

# Ejemplo de uso
texto_t = texto_limpio_t
texto_m = texto_limpio_m

# Usar la función para filtrar ambos textos
texto_filtrado_t, texto_filtrado_m = tokens_dos_textos(texto_t, texto_m, idioma='english')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Encontrar la intersección de los tokens entre ambos DataFrames
tokens_comunes = tokens_t.intersection(tokens_m)

# Crear un DataFrame con los tokens comunes y sus frecuencias en ambos DataFrames
df_comunes_t = df_frecuencias_filtrado_t[df_frecuencias_filtrado_t['token'].isin(tokens_comunes)]
df_comunes_m = df_frecuencias_filtrado_m[df_frecuencias_filtrado_m['token'].isin(tokens_comunes)]

# Unir los dos DataFrames por el token
df_comunes = pd.merge(df_comunes_t[['token', 'conteo']], df_comunes_m[['token', 'conteo']], on='token', how='inner', suffixes=('_t', '_m'))

# Graficar las frecuencias de los tokens comunes
plt.figure(figsize=(10, 6))
df_comunes.plot(kind='bar', x='token', y=['conteo_t', 'conteo_m'], color=['skyblue', 'lightcoral'])

# Personalizar el gráfico
plt.xlabel('Tokens', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Asegurarse de que el layout esté ajustado antes de guardar
plt.tight_layout()

# Guardar el gráfico como una imagen
plt.savefig('Comun.png', format='png', dpi=300)  # DPI alto para calidad
plt.show()
```
