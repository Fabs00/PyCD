```python
# URL de la publicación específica
url = "https://www.reddit.com/r/AskReddit/comments/pcfrxa/exantivaxxers_of_reddit_what_made_you_change_your/"

# Obtener la publicación usando la URL
submission = reddit.submission(url=url)

# Cargar todos los comentarios (incluyendo respuestas)
submission.comments.replace_more(limit=None)
# Función para extraer comentarios con fecha
def get_comments(comment, level=0):
    comments_data.append({
        'author': comment.author.name if comment.author else "Anonymous",
        'text': comment.body,
        'score': comment.score,
        'date': datetime.utcfromtimestamp(comment.created_utc).strftime('%Y-%m-%d %H:%M:%S'),
        'depth': level
    })
    for reply in comment.replies:
        get_comments(reply, level + 1)

# Obtener comentarios y sus respuestas
comments_data = []
for top_level_comment in submission.comments:
    get_comments(top_level_comment)

# Convertir los datos a un DataFrame de Pandas
df_comments = pd.DataFrame(comments_data)

# Mostrar los primeros comentarios con fecha
print(df_comments[['author', 'text', 'date','score','depth']].head())

# Guardar en CSV
df_comments.to_csv('reddit_comments_with_dates.csv', index=False)

print("Comentarios guardados correctamente.")
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Descargar recursos de NLTK (solo la primera vez)
nltk.download('punkt')
nltk.download('stopwords')

# Definir la función de tokenización
def tokenizar(txt):
    if pd.isna(txt):  # Manejo de valores nulos
        return []
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())  # Eliminar signos de puntuación y convertir a minúsculas
    stop_words = set(stopwords.words('english'))  # Stopwords en inglés (cambiar a 'spanish' si es necesario)
    tokens = word_tokenize(texto_limpio)  # Tokenizar el texto
    texto_filtrado = [word for word in tokens if word not in stop_words]  # Filtrar stopwords
    return texto_filtrado

# Aplicar la función a la columna de comentarios en tu DataFrame
df_comments['tokens'] = df_comments['text'].apply(tokenizar)

# Mostrar los primeros resultados
print(df_comments[['text', 'tokens']].head())
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Descargar recursos de NLTK (si aún no están disponibles)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Inicializar el lematizador
lemmatizer = WordNetLemmatizer()

# Función para limpiar, tokenizar y lematizar
def procesar_texto(txt):
    if pd.isna(txt):  # Manejo de valores nulos
        return []

    # Preprocesamiento: eliminar puntuación y poner en minúsculas
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())

    # Tokenización
    tokens = word_tokenize(texto_limpio)

    # Eliminar stopwords
    stop_words = set(stopwords.words('english'))
    tokens_filtrados = [word for word in tokens if word not in stop_words]

    # Lematización
    tokens_lematizados = [lemmatizer.lemmatize(word) for word in tokens_filtrados]

    return tokens_lematizados

# Aplicar la función a la columna de comentarios del DataFrame
df_comments['lemmatized_tokens'] = df_comments['text'].apply(procesar_texto)

# Mostrar los primeros resultados
print(df_comments[['text', 'lemmatized_tokens']].head())
```

```python
import pandas as pd
from collections import Counter

# Suponiendo que tu columna de tokens se llama 'tokens' y contiene listas de palabras
all_tokens = df_comments['tokens'].explode().dropna().tolist()  # Convertir listas en una sola lista plana

# Contar la frecuencia de las palabras
word_counts = Counter(all_tokens)

# Convertir a un DataFrame para mejor visualización
df_word_counts = pd.DataFrame(word_counts.items(), columns=['word', 'count'])

# Ordenar por frecuencia descendente
df_word_counts = df_word_counts.sort_values(by='count', ascending=False)

# Mostrar las 10 palabras más frecuentes
print(df_word_counts.head(20))
```

```python
def perfil_sentimientos(df):
    print(df.reset_index().loc[0,'author'])
    print("=" * 12)
    print(f"Positivos: {round(100 * np.mean(df.sentimiento > 0), 2)}")
    print(f"Neutros  : {round(100 * np.mean(df.sentimiento == 0), 2)}")
    print(f"Negativos: {round(100 * np.mean(df.sentimiento < 0), 2)}")
    print(" ")
```

```python
for author, df in df_sentimiento.groupby("author"):
    perfil_sentimientos(df)
```
